# What was the best month for sales? How much was earned that month?

# Fetching month from date

```python
df['Month'] = df['Order Date'].dt.month
df.head()
```

```python
df['Month_name'] = df['Order Date'].dt.month_name()
df.head()
```

```python
mdf=df.groupby('Month_name').agg(totalsale=('Amount','sum'),avgsale=('Amount','mean'))
mdf=mdf.reset_index()
mdf=mdf.sort_values(by='totalsale',ascending=False)
mdf
```

```python
mdf.to_csv('SalesData/TopMonth.csv',index=False)
```

# Visulaisation 1

```python
plt.figure(figsize=(12,5))
plt.bar(mdf['Month_name'],mdf['totalsale'],color='orange')
plt.title('Sales')
plt.xlabel('Month')
plt.ylabel('TotalSale')
plt.xticks(rotation=45,color='grey')
plt.yticks()
plt.grid()
plt.savefig('SalesData/Monthlysales.png')
```

```python
tdf=df.groupby(['Month','Month_name']).agg(totalsale=('Amount','sum'))
tdf=tdf.reset_index()
tdf
```

```python
plt.figure(figsize=(12,5))
plt.plot(tdf['Month_name'],tdf['totalsale'],color='orange')
plt.title('Sales',color ='green')
plt.xlabel('Month')
plt.ylabel('TotalSale')
plt.xticks(rotation=45,color='brown')
plt.yticks()
plt.grid()
plt.savefig('SalesData/Monthwise_sales.png')
```

```python
plt.pie(tdf['totalsale'],labels=tdf['Month_name'],autopct='%.1f%%',radius=1.5)
plt.savefig('SalesData/Monthwise_pie_sales.png')
```

# What city sold the most product ?

```python
s=df['Purchase Address'].str.split(',')
s
```

```python
def city(l):
    return l[1]

s =df['Purchase Address'].str.split(',')
df['city'] =s.apply(city)
df
```

```python
d=df['Purchase Address'].str.split(',',expand=True)
df['city']=d[1]
df
```

```python
ndf=df.groupby('city').agg(totalquantity=('Quantity Ordered','sum'),totalsale=('Amount','sum'))
ndf.to_csv('SalesData/city.csv')
```

```python
ndf=ndf.reset_index()
ndf=ndf.sort_values(by='totalsale',ascending=False)
ndf
```

```python
ndf=ndf.reset_index(drop=True)
ndf
```

# City -graph

```python
plt.figure(figsize=(12,5))
plt.bar(ndf['city'],ndf['totalsale'])
plt.xlabel('City')
plt.ylabel('TotalSale (in 10 lakhs)')

plt.savefig('SalesData/totalsales.png')
```
